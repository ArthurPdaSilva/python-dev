```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

path = "files/apostila.pdf"
loader = PyPDFLoader(path)
docs = loader.load()

recur_char_split = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", " ", ""]
)
split_docs = recur_char_split.split_documents(docs)

directory = "files/chroma_vectorstore"

embeddings = OpenAIEmbeddings()
vector_store = Chroma.from_documents(
    split_docs, embeddings, persist_directory=directory
)
```

```python
vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory=directory
)

question = "Principais metodos de manipulação de strings?"
docs = vector_store.similarity_search(question, k=5)

```
